# 2: Load and Train YOLOv5 Model

```python
from ultralytics import YOLO

# Load model (keep this, it's correct)
model = YOLO("yolov5s.pt")

# Optimized training config for GTX 1650 Ti
results = model.train(
    data="vehicle_dataset/data.yaml",
    epochs=50,
    imgsz=512,  # lower than 640 → fits VRAM better
    batch=8,  # safe starting point for 4GB GPU
    device="cuda",  # USE GPU
    workers=2,  # don’t overload your CPU
    cache=False,  # avoid RAM overload
    amp=True,  # mixed precision → faster + less VRAM
)

print("Training complete!")
print(results)
```

# 3: Validate the Trained Model

```python
val_results = model.val(data="vehicle_dataset/data.yaml", split="test")
print("Validation results:")
print(val_results)
```

# 4: Predict on Sample Images (Optional)

```python
test_images = [
    "vehicle_dataset/images/test/MVI_39031_000010.jpg",  # Replace with actual image paths if needed
    "vehicle_dataset/images/test/MVI_39051_000020.jpg",
]

for img_path in test_images:
    if os.path.exists(img_path):
        results = model.predict(source=img_path, save=True, conf=0.5)
        print(f"Predictions for {img_path}:")
        print(results)
    else:
        print(f"Image {img_path} not found.")
```
